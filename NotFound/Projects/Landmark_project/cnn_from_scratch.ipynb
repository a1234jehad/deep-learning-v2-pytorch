- How does your code resize the images (by cropping, stretching, etc)?  What size did you pick for the input tensor, and why?

<img src="static_images/icons/noun-answer-3361020.png" alt=">" style="width:25px"/> **Answer**: My code first resizes the image to 256 and then crops to 224. I picked 224 as the input size because it is the recommended input size for using pytorch's pre-trained models. I did decide to augment the dataset via RandAugment, a typical set of augmentations for natural images. I added this augmentation with the goal of improving my model's robustness, thus improving test accuracy.

Resizing Method: The transforms.Resize(256) command is used to resize the image. This transform resizes the smaller edge of the image to 256 pixels. If the image is not square, the aspect ratio of the image is maintained, meaning only the smaller dimension (height or width) will be resized to 256 pixels, while the other dimension will be scaled to maintain the aspect ratio.

Cropping: After resizing, transforms.CenterCrop(224) is used. This crop operation cuts out a 224x224 pixel square from the center of the image. The choice of 224x224 pixels is a standard in many deep learning models, particularly those trained on the ImageNet dataset.

Size Choice for Input Tensor (224x224): The size of 224x224 pixels for the input tensor is a widely used standard, primarily because it strikes a balance between having enough detail in the image and not having too large an image that would require more computational resources. This size is especially common in models like the various versions of the VGG and ResNet architectures, which were originally designed and trained with ImageNet data. Using this standard size allows for easy use of pre-trained models and weights.

Why These Choices?

Resizing: Ensures that all images fed into the model have a consistent size, which is necessary since most deep learning models expect input data of a uniform size.
Center Cropping: Focuses on the center of the image, which in many cases is where the subject of the photo is located. This can help in situations where the subject is in the center, although it may cut off relevant parts of the image if the subject is not centrally located.
224x224 Pixels: Provides a good balance between image detail and computational efficiency. Larger images would provide more detail but at the cost of increased computational requirements, while smaller images would be faster to process but might lack necessary details.


Decision to Augment the Dataset: Yes, the dataset is augmented, but only for the training set. Data augmentation is a technique used to increase the diversity of the training data without actually collecting new data. This is achieved by applying various transformations to the existing data, which helps the model generalize better and reduces the risk of overfitting.

Method of Augmentation - AutoAugment: The AutoAugment transformation is applied. AutoAugment automatically selects and applies a series of image transformations, such as translations, flips, rotations, color adjustments, and more. The specific transformations and their intensities are determined based on the policy selected. In the provided code, AutoAugmentPolicy.IMAGENET is suggested, which applies transformations found to be effective for the ImageNet dataset. These transformations are diverse and have been optimized through extensive research.

Types of Augmentations:

Translations, Flips, Rotations: These are common augmentations that shift, mirror, or rotate the image, respectively. They help the model learn to recognize objects regardless of their position, orientation, or symmetry in the image.
Color Adjustments: Changes in brightness, contrast, saturation, and hue. These help the model become robust to variations in lighting and color, which is especially useful if the model is expected to perform in different lighting conditions or with varying camera qualities.
Other complex augmentations may also be included depending on the specific policy of AutoAugment.

Why Augment?: Augmentation is particularly useful when the available training data is limited or when the model needs to be robust to various transformations (like in real-world scenarios). It helps in effectively increasing the dataset size and variation, leading to better model generalization.

<img src="static_images/icons/noun-question-mark-869751.png" alt="?" style="width:25px"/> **Question**: Outline the steps you took to get to your final CNN architecture and your reasoning at each step.  

<img src="static_images/icons/noun-answer-3361020.png" alt=">" style="width:25px"/> __Answer:__ I decided to use 4 convolutional layers so that my model could be sufficiently expressive. I used dropout layers to reduce my model's tendency to overfit the training data. I made my model output a 50-dimensional vector to match with the 50 available landmark classes.


Initial Convolutional Layers (Increasing Channels):

Layers: Four convolutional layers with increasing numbers of channels (16, 32, 64, 128). Each followed by batch normalization and ReLU activation.
Reasoning: The increasing number of channels allows the model to learn more complex features at each layer. The first layers capture basic features like edges, and as we move deeper, the features become more complex.

Batch Normalization:
Applied after each convolutional layer.
Reasoning: Improves training speed and stability by normalizing the outputs of the previous layer. Helps in reducing internal covariate shift.

ReLU Activation:
Used after each convolutional layer and fully connected layer.
Reasoning: Introduces non-linearity into the model, allowing it to learn more complex patterns. ReLU is preferred for its simplicity and efficiency.

Max Pooling:
Applied after each ReLU in the convolutional stages.
Reasoning: Reduces the spatial dimensions (width and height) of the input volume for the next convolutional layer. It helps in reducing computation and controlling overfitting by providing an abstracted form of the representation.

Flattening:
Transforming the 2D feature maps into a 1D vector.
Reasoning: Necessary to transition from convolutional layers to fully connected layers.

Fully Connected Layers:
One hidden layer with 512 units and the output layer with num_classes units.
Reasoning: The dense layer with 512 units allows the network to learn a global representation of the features extracted by the convolutional layers. The final output layer matches the number of classes for classification.

Dropout:
Applied after the first fully connected layer.
Reasoning: Helps in reducing overfitting by randomly setting a fraction (dropout rate) of input units to 0 at each update during training time. It forces the network to learn redundant representations and improves generalization.

Output Layer:
Linear layer with output size num_classes.
Reasoning: The final layer outputs a logit for each class. This is used with a loss function like Cross-Entropy to perform multi-class classification.

In summary, the architecture is a typical convolutional neural network with layers designed to progressively extract more complex features, normalize data, introduce non-linearity, and reduce dimensions, followed by dense layers for classification. The use of dropout helps in mitigating overfitting. The architecture is structured to be adaptable to different numbers of output classes (num_classes) and varying degrees of regularization (dropout).

Allright, good job getting here! Now it's time to see if all our hard work pays off. In the following cell we will train your model and validate it against the validation set.

Let's start by defining a few hyperparameters. Feel free to experiment with different values and try to optimize your model:

````python
batch_size = 128        # size of the minibatch for stochastic gradient descent (or Adam)
valid_size = 0.2       # fraction of the training data to reserve for validation
num_epochs = 100        # number of epochs for training
num_classes = 50       # number of classes. Do not change this
dropout = 0.5          # dropout for our model
learning_rate = 0.001  # Learning rate for SGD (or Adam)
opt = 'sgd'            # optimizer. 'sgd' or 'adam'
weight_decay = 0   # regularization. Increase this to combat overfitting
````

````python
from src.data import get_data_loaders
from src.train import optimize
from src.optimization import get_optimizer, get_loss
from src.model import MyModel

# get the data loaders using batch_size and valid_size defined in the previous
# cell
# HINT: do NOT copy/paste the values. Use the variables instead
data_loaders = get_data_loaders(batch_size=batch_size, valid_size=valid_size)

# instance model MyModel with num_classes and drouput defined in the previous
# cell
model = MyModel(num_classes=num_classes, dropout=dropout)

# Get the optimizer using get_optimizer and the model you just created, the learning rate,
# the optimizer and the weight decay specified in the previous cell
optimizer = get_optimizer(model=model,learning_rate= learning_rate,optimizer= opt, weight_decay=weight_decay)

# Get the loss using get_loss
loss = get_loss()

optimize(
    data_loaders,
    model,
    optimizer,
    loss,
    n_epochs=num_epochs,
    save_path="checkpoints/best_val_loss.pt",
    interactive_tracking=True
)
````
